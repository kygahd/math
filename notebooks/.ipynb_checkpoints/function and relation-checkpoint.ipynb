```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_graph(x_values, y_values):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    # plot the function
    plt.plot(x_values, y_values, 'r')

    # show the plot
    plt.show()
```

```python
def piecewise_func(x):
    if x < 2:
        return 2 * x + 2  # x < 2: y=2x+2
    elif x >=2:
        return 2 ** x  # x >= 2: y=2 ^ x
    

# x_coordinates = [-10, -5, 0, 5, 10]
x_coordinates = np.linspace(-5,5,100)
y_coordinates = [piecewise_func(x) for x in x_coordinates]

# print(f'x_coordinates: {x_coordinates}')
# print(f'y_coordinates: {y_coordinates}')

plot_graph(x_coordinates, y_coordinates)
```

function 1: $y = 2 ^ x$

```python
def func_one(x):
    # y = 2 ^ x
    return 2 ** x

x_coordinates = np.linspace(-5,5,100)
y_coordinates = [func_one(x) for x in x_coordinates]

plot_graph(x_coordinates, y_coordinates)
```
